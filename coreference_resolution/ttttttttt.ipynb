```python
import os
def spans(txt):
    tokens=nltk.word_tokenize(txt)
    offset = 0
    for token2 in tokens:
        add = 0
        if token2=="''" or token2 =='``':
             token = '\"'
             add = 1
        else:
            token = token2
        offset = txt.find(token, offset)
        yield token, offset, offset+len(token)
        offset += len(token)

path = '../a_dataset_rucoref/rucoref_texts/GG2/'
files = os.listdir(path)
files.sort(key=lambda s: int(s))
with open('../a_dataset_rucoref/rucoref_texts/my_token2.txt', 'w') as fw:
    fw.write('doc_id	shift	length	token	lemma	gram')
    fw.write('\n')

    for file in files:
        with open(os.path.join(path, file), 'r') as f:
            text = f.read()
            for token in spans(text):
                fw.write('{}\t{}\t{}\t{}\t{}\t{}'.format(1000+int(file), token[1], len(token[0]), token[0], token[0], 'SENT' if token[0] in ['.','!','?'] else 'R'))
                fw.write('\n')
```
